```python
#PIE CHART




''' 
PIE chart moves anti-clockwise
'''
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True,explode=[0, 0.2, 0.1, 0, 0],startangle=180)
plt.show()
```
